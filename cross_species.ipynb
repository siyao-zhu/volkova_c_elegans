### Now follow the paper method to categorize the COSMIC 83 into 14 bin worm scheme

```python
import numpy as np
import pandas as pd
from itertools import chain

# concatenate any number of iterables of ints into one int array
def idx(*items) -> np.ndarray:
    if not items:
        return np.array([], dtype=int)
    return np.fromiter(chain.from_iterable(items), dtype=int)


def regroup_cosmic_id83_to_14(id83: pd.DataFrame) -> pd.DataFrame:
    if 'Type' in id83.columns:
        id83 = id83.drop(columns=['Type'])
        
    # follow exact rules in the paper => according to their r markdown files
    rows = [
        idx(range(3,7), range(9,13)),                                               # 1) D.1.rep
        idx(range(1,3), range(7,9)),                                                # 2) D.1.nonrep
        idx(range(27,31), range(33,37), range(39,43)),                              # 3) D.2.5.rep
        idx(range(25,27), range(31,33), range(37,39), range(73,79)),                # 4) D.2.5.nonrep
        idx(range(43,49), range(79,84)),                                            # 5) D.5.50
        np.array([], int),                                                          # 6) D.50.400 = 0
        np.array([], int),                                                          # 7) DI.small = 0
        np.array([], int),                                                          # 8) DI.large = 0
        idx(range(15,19), range(21,25)),                                            # 9) I.1.rep
        idx(range(13,15), range(19,21)),                                            # 10) I.1.nonrep
        idx(range(51,55), range(57,61), range(63,67)),                              # 11) I.2.5.rep
        idx(range(49,51), range(55,57), range(61,63)),                              # 12) I.2.5.nonrep
        idx(range(67,73)),                                                          # 13) I.5.50
        np.array([], int),                                                          # 14) I.50.400 = 0
    ]

    out = []
    for r in rows:
        if r.size == 0:
            out.append(np.zeros((id83.shape[1],), float))
        else:
            block = id83.iloc[(r-1)].values if (id83.index.min()==1) else id83.iloc[r].values
            out.append(block.sum(axis=0))
    M = np.vstack(out)
    return pd.DataFrame(M, index=order14, columns=id83.columns)
```

### Then show signature after converting to human genome

```python
# 1. load the c.elegans signature
df = pd.read_csv("table2A_indel14_genotypes_mean.csv")

if "class" not in df.columns:
    df = df.rename(columns={df.columns[0]: "class"})

missing = [c for c in order14 if c not in df["class"].tolist()]
if missing:
    raise ValueError(f"Missing expected rows in Table 2A: {missing}")

df = df.set_index("class").loc[order14].reset_index()

geno_cols = [c for c in df.columns if c != "class"]

# (14 x n_genotypes)
M = df[geno_cols].to_numpy(dtype=float)

eps = 1e-8

# normalization => mlh.1 is really strange
col_sums = np.clip(M.sum(axis=0, keepdims=True), eps, None)
M = M / col_sums
```

```python
# 2. S_humanized[c] = S_worm[c] / BG_worm[c] * BG_human[c]
eps = 1e-8
bw = np.clip(np.asarray(ce_bg14, dtype=float).reshape(-1,1), eps, None)     # (14,1)
bh = np.clip(np.asarray(human_bg14, dtype=float).reshape(-1,1), eps, None)  # (14,1)

conv = (M / bw) * bh
conv /= np.clip(conv.sum(axis=0, keepdims=True), eps, None)
```

```python
df_before = pd.DataFrame(M, index=order14, columns=geno_cols).reset_index().rename(columns={'index':'class'})
df_before
```

```python
df_after = pd.DataFrame(conv, index=order14, columns=geno_cols).reset_index().rename(columns={"index":"class"})
df_after
```

```python
def cosine_sim(A: np.ndarray, B: np.ndarray, eps: float = 1e-8) -> np.ndarray:
    A = A.astype(float, copy=False); B = B.astype(float, copy=False)
    A /= np.clip(np.linalg.norm(A, axis=1, keepdims=True), eps, None)
    B /= np.clip(np.linalg.norm(B, axis=1, keepdims=True), eps, None)
    return A @ B.T

Y_before = M.T
Y_after = conv.T
```

```python
# Make columns numeric and set a 1-based integer index so the function uses (r-1) in regroup_cosmic_id83_to_14
cosmic_id_num = cosmic_id.apply(pd.to_numeric, errors="coerce")
cosmic_id_num.columns = cosmic_id_num.columns.str.strip()
cosmic_id_num.index = np.arange(1, cosmic_id_num.shape[0] + 1)

cosmic14 = regroup_cosmic_id83_to_14(cosmic_id_num)
Y_cosmic14 = cosmic14.T.values
cosmic_names = cosmic14.columns.tolist()
```

```python
# Cosine similarities
sim_before = cosine_sim(Y_before, Y_cosmic14)
sim_after = cosine_sim(Y_after, Y_cosmic14)

# Top-1 match per genotype
best_b = sim_before.argmax(axis=1)
best_a = sim_after.argmax(axis=1)
```

```python
genotypes = list(df_after.columns)[1:]

summary = pd.DataFrame({
    "genotype":   genotypes,
    "top_before": [cosmic_names[j] for j in best_b],
    "sim_before": [float(sim_before[i, best_b[i]]) for i in range(len(genotypes))],
    "top_after":  [cosmic_names[j] for j in best_a],
    "sim_after":  [float(sim_after[i, best_a[i]])  for i in range(len(genotypes))],
})
summary["delta"]   = summary["sim_after"] - summary["sim_before"]
summary["changed"] = summary["top_before"] != summary["top_after"]

summary
```

```python
import numpy as np
import matplotlib.pyplot as plt
import musical

# Already have:
#   df_before, df_after  (rows=14 classes + 'class' col; cols=genotypes)
#   cosmic14 (14 x K DataFrame; collapsed from ID83 as earlier)
#   sim_before, sim_after (N_genotypes x K) cosine-sim matrices

def _vec14_from_df(df, genotype, order14):
    df2 = df.set_index("class").loc[order14]
    return df2[genotype].to_numpy()

def plot_signature14_quad(
    genotype, df_before, df_after, cosmic14, summary, order14=order14
):
    """
    Panels (2x2): BEFORE (worm) | AFTER (humanized)
                  Top COSMIC BEFORE | Top COSMIC AFTER
    """
    row = summary.loc[summary["genotype"] == genotype].iloc[0]
    
    v_before = _vec14_from_df(df_before, genotype, order14)
    v_after  = _vec14_from_df(df_after,  genotype, order14)

    v_cos_b = cosmic14[row["top_before"]].to_numpy()
    v_cos_a = cosmic14[row["top_after"]].to_numpy()
    if v_cos_b.sum() > 0: v_cos_b = v_cos_b / v_cos_b.sum()
    if v_cos_a.sum() > 0: v_cos_a = v_cos_a / v_cos_a.sum()

    x = np.arange(len(order14))
    fig, axes = plt.subplots(2, 2, figsize=(14, 8), sharey=True)

    axes[0,0].bar(x, v_before); axes[0,0].set_title(f"{genotype} — BEFORE")
    axes[0,1].bar(x, v_after);  axes[0,1].set_title(f"{genotype} — AFTER (humanized)")

    axes[1,0].bar(x, v_cos_b)
    axes[1,0].set_title(f"Top COSMIC BEFORE: {row['top_before']}\ncosine={row['sim_before']:.3f}")

    axes[1,1].bar(x, v_cos_a)
    axes[1,1].set_title(f"Top COSMIC AFTER: {row['top_after']}\ncosine={row['sim_after']:.3f}")

    for ax in axes.ravel():
        ax.set_xticks(x); ax.set_xticklabels(order14, rotation=90)
        ax.set_ylabel("Normalized weight")
    plt.tight_layout()
    plt.show()

def plot_signature14_triptych(
    genotype, df_before, df_after, cosmic14, summary, order14=order14, which_cosmic="after"
):
    """
    Panels: BEFORE (worm) | AFTER (humanized) | top COSMIC (from summary['top_after' or 'top_before'])
    which_cosmic: 'after' or 'before'
    """
    row = summary.loc[summary["genotype"] == genotype].iloc[0]
    sig_name = row["top_after"] if which_cosmic == "after" else row["top_before"]
    v_cosmic = cosmic14[sig_name].to_numpy()
    v_cosmic = v_cosmic / v_cosmic.sum() if v_cosmic.sum() > 0 else v_cosmic

    v_before = _vec14_from_df(df_before, genotype, order14)
    v_before = v_before / v_before.sum() if v_before.sum() > 0 else v_before

    v_after  = _vec14_from_df(df_after,  genotype, order14)

    x = np.arange(len(order14))
    fig, axes = plt.subplots(1, 3, figsize=(14, 4), sharey=True)

    axes[0].bar(x, v_before); axes[0].set_title(f"{genotype} — BEFORE")
    axes[1].bar(x, v_after);  axes[1].set_title(f"{genotype} — AFTER (humanized)")
    lbl = f"Top COSMIC ({which_cosmic}): {sig_name}"
    if which_cosmic == "after":
        lbl += f"\ncosine={row['sim_after']:.3f}"
    else:
        lbl += f"\ncosine={row['sim_before']:.3f}"
    axes[2].bar(x, v_cosmic); axes[2].set_title(lbl)

    for ax in axes:
        ax.set_xticks(x); ax.set_xticklabels(order14, rotation=90)
        ax.set_ylabel("Normalized weight")
    plt.tight_layout(); plt.show()
```

```python
import os, glob, re, pickle
import numpy as np

BG_DIR = "/Users/zhusiyao/Desktop/Hu Jin lab/Indel mutational signatures/pkl to pict/c.elegans/cagan_cross_species/background_vc"

bg94 = {}

for path in sorted(glob.glob(os.path.join(BG_DIR, "*.pkl"))):
    with open(path, "rb") as f:
        vec = pickle.load(f)
    arr = np.asarray(vec)

    # Collapse to one vector of length 94
    if arr.ndim == 2 and arr.shape[1] == 94:
        vec94 = arr.sum(axis=0)
    elif arr.ndim == 1 and arr.shape[0] == 94:
        vec94 = arr
    else:
        raise ValueError(f"{os.path.basename(path)} has unexpected shape {arr.shape} (need [*,94] or [94])")

    # Make a clean species key from filename ("ferret_indel94_windows.pkl" -> "ferret")
    base = os.path.basename(path)
    name = re.sub(r'\.pkl$', '', base)
    name = re.sub(r'_indel(?:83|94).*$', '', name)  # drop suffixes like _indel94_...
    name = name.replace('_', ' ').strip()

    bg94[name] = np.asarray(vec94, dtype=float)

print(f"Loaded {len(bg94)} backgrounds:")
for sp, v in bg94.items():
    print(f"  {sp}: shape={v.shape}, sum={v.sum():.0f}")
```

```python
import pandas as pd
def ID94_to_ID83(x):
    x = pd.Series.copy(x)
    x.loc['INS.repeats.2.0'] = x.loc['INS.repeats.2.0'] + x.loc['INS.MH.2.1']
    x.loc['INS.repeats.3.0'] = x.loc['INS.repeats.3.0'] + x.loc['INS.MH.3.1'] + x.loc['INS.MH.3.2']
    x.loc['INS.repeats.4.0'] = x.loc['INS.repeats.4.0'] + x.loc['INS.MH.4.1'] + x.loc['INS.MH.4.2'] + x.loc['INS.MH.4.3']
    x.loc['INS.repeats.5+.0'] = x.loc['INS.repeats.5+.0'] + x.loc['INS.MH.5+.1'] + x.loc['INS.MH.5+.2'] + x.loc['INS.MH.5+.3'] + x.loc['INS.MH.5+.4'] + x.loc['INS.MH.5+.5+']
    return pd.Series.copy(x.iloc[0:83])
```

```python
EPS = 1e-8

def convert_bg94_to_id83(bg94_dict, normalize=True, rename_columns=True):
    rows = []
    names = []
    for sp, arr94 in bg94_dict.items():
        ser94 = pd.Series(np.asarray(arr94, dtype=float), index=indel_types_94_str, name=sp)
        ser83 = ID94_to_ID83(ser94)

        vec83 = ser83.iloc[:83].to_numpy(dtype=float)
        if normalize:
            vec83 = np.maximum(vec83, EPS)
            vec83 = vec83 / vec83.sum()

        rows.append(vec83)
        names.append(sp)

    opp83 = pd.DataFrame(rows, index=names)
    if rename_columns:
        opp83.columns = indel_types_83_str
    return opp83

opp83 = convert_bg94_to_id83(bg94, normalize=True)

print("Converted backgrounds (ID-83):")
print(opp83.shape)
opp83.head()

# opp83.to_csv("background_vc/ID83_opportunities_from_pkls.tsv", sep="\t")
```

```python
def norm_name(s: str) -> str:
    return str(s).replace("_", " ").strip().lower()

df_tmp = df.copy()
df_tmp.index = df_tmp.index.map(norm_name)

opp83_tmp = opp83.copy()
opp83_tmp.index = opp83_tmp.index.map(norm_name)

# Keep only species present in BOTH
common = sorted(set(df_tmp.index) & set(opp83_tmp.index))
missing_in_bg = sorted(set(df.index) - set(opp83_tmp.index))
missing_in_df = sorted(set(opp83_tmp.index) - set(df.index))

dfc = df_tmp.loc[common, indel_types_83_str].astype(float)
oppc = opp83_tmp.loc[common, indel_types_83_str].astype(float)

# Background-adjust: y_adj ∝ y / opp ; then renormalize to sum=1
EPS = 1e-8
adj = dfc.div(oppc.clip(lower=EPS), axis=0)
adj = adj.div(adj.sum(axis=1).replace(0, 1), axis=0)

# 'adj' has background-adjusted per-species ID-83 spectra (freq)
for sp, row in adj.iterrows():
    vec = row.to_numpy(float)
    musical.sigplot_bar(
        vec,
        sig_type='Indel83',
        title=f"{sp.title()} — ID-83 spectrum (background-adjusted)",
        norm=True,
        xlabel="Indel Mutation Type",
        ylabel="Normalized Count",
        figsize=(10, 4),
        tick_fontsize=8,
        label_fontsize=12,
        rotation=90,
        xticklabels=True
    )
    plt.show()
```

### Switch to adjusting the bg per groups (5 groups in total)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from musical import plot as musical_plot

def _norm(s): return str(s).replace("_"," ").strip().lower()
df2 = df.copy()
opp2 = opp83.copy()
df2.index  = df2.index.map(_norm);  opp2.index = opp2.index.map(_norm)
df2 = df2.loc[sorted(set(df2.index) & set(opp2.index))]
opp2 = opp2.loc[df2.index]
df2 = df2.astype(float)
opp2 = opp2.astype(float)

# normazlied each row => sum to 1
opp2 = opp2.div(opp2.sum(axis=1).replace(0,1), axis=0)

# Robust background-adjustment
# 5 groups: [0:12]=1bp DEL, [12:24]=1bp INS,
# [24:48]=>1bp DEL@repeats, [48:72]=>1bp INS@repeats, [72:83]=DEL.MH
GROUPS = [(0,12),(12,24),(24,48),(48,72),(72,83)]
EPS = 1e-8 # avoid divide-by-zero
ALPHA = 1e-6 # prevents zero bins from exploding
BETA  = 0.02 # 2% mix towards uniform within each group (stabilizes tiny opp)

def adjust_row(y, o):
    y = np.asarray(y, float)
    o = np.asarray(o, float)
    y_s = y + ALPHA
    o_s = o + ALPHA
    out = np.zeros_like(y_s)
    for a,b in GROUPS:
        yg = y_s[a:b]
        og = o_s[a:b]
        # mix opp with uniform inside group to avoid tiny zeros dominating
        if BETA > 0:
            k = len(og)
            og = (1-BETA)*og + BETA*(og.sum()/k)
        rg = yg / np.maximum(og, EPS)
        # preserve the group's original mass (sum of y within the group)
        gm = yg.sum()
        if rg.sum() > 0:
            # keeps the total within-group the same
            # only the distribution inside the group changes after correction
            rg = rg / rg.sum() * gm
        out[a:b] = rg
    # final renormalization
    s = out.sum()
    return out / s if s > 0 else out

adj = pd.DataFrame(
    [adjust_row(df2.loc[sp].values, opp2.loc[sp].values) for sp in df2.index],
    index=df2.index, columns=df2.columns
)

# Plot an example
sp = 'colobus'
for title, mat in [("ID-83 spectrum", df2), ("ID-83 spectrum (background-adjusted)", adj)]:
    vec = mat.loc[sp].to_numpy(float)
    musical_plot.sigplot_bar(
        vec, sig_type='Indel83', title=f"{sp.title()} — {title}",
        norm=True, xlabel="Indel Mutation Type", ylabel="Normalized Count",
        figsize=(10,4), tick_fontsize=8, label_fontsize=12, rotation=90, xticklabels=True
    )
    plt.show()

# quantify “more similar” via cosine similarity pre vs post
from scipy.spatial.distance import pdist, squareform
pre  = 1 - squareform(pdist(df2.values,  metric='cosine'))
post = 1 - squareform(pdist(adj.values, metric='cosine'))
print("Mean pairwise cosine similarity: pre =", pre[np.triu_indices_from(pre,1)].mean(),
      " post =", post[np.triu_indices_from(post,1)].mean())
```
